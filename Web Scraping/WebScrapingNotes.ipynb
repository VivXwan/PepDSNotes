```python
name = []
ram = []
rom = []
rating = []
discount = []
for mobile in mobiles:
  name.append(mobile.find('div', class_ = '_4rR01T').text)
  price.append(mobile.find('div', class_ = '_30jeq3 _1_WHN1').text)
  specs = mobile.find('li', class_ = 'rgWa7D')
  ram.append(specs.text.split("|")[0].strip())
  rom.append(specs.text.split("|")[1].strip())
  rating.append(mobile.find('div',class_ = '_3LWZlK').text)
  discount.append(mobile.find('div',class_ = '_3Ay6Sb').text)
```

```python
print(name,price,ram,rom,rating,discount,sep="\n")
```

```python
import pandas as pd
```

```python
zipped = list(zip(name,ram,rom,rating,price,discount))
pd.DataFrame(zipped,columns = ["Name","RAM","ROM","Rating","Price","Discount"])
```

```python
from selenium import webdriver
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
options = webdriver.ChromeOptions()

options.add_argument('--headless')
options.add_argument('--no-sandbox')

driver = webdriver.Chrome('chromedriver',options = options)
```

```python
import time
driver.get("https://summerofcode.withgoogle.com/archive/2021/organizations")
time.sleep(2)
```

```python
page = driver.page_source
```

```python
page
```

```python
soup = bs(page,"html.parser")
```

```python
soup.title # Is the Required page since the title is same.
```

```python
cards = soup.findAll("div",class_="card")
```

```python
card = cards[0]
```

```python
card.find("div",class_="name").text.strip()
```

```python
"https://summerofcode.withgoogle.com"+card.a['href']
```

```python
name = []
url = []
for card in cards:
  name.append(card.find("div",class_="name").text.strip())
  url.append("https://summerofcode.withgoogle.com"+card.a['href'])
```

```python
cards = pd.DataFrame({"Name":name,"URL":url})
```

```python
cards
```

```python
for url in cards['URL']:
  driver.get(url)
  org_page = driver.page_source
```
